```python
PATH_DISCOGS_XML = '/home/rute/Code/teste-musica-portuguesa/discogs data dumps/'
FILENAME_DISCOGS = 'discogs_20210101_releases.xml'
FILENAME_RELEASES = 'releases.csv'
ENCODING = "utf-8"
```

```python
# Nicely formatted time string
def hms_string(sec_elapsed):
    h = int(sec_elapsed / (60 * 60))
    m = int((sec_elapsed % (60 * 60)) / 60)
    s = sec_elapsed % 60
    return "{}:{:>02}:{:>05.2f}".format(h, m, s)
```

```python
def strip_tag_name(t):
    t = elem.tag
    idx = k = t.rfind("}")
    if idx != -1:
        t = t[idx + 1:]
    return t
```

```python
pathWikiXML = os.path.join(PATH_DISCOGS_XML, FILENAME_DISCOGS)
pathMasters = os.path.join(PATH_DISCOGS_XML, FILENAME_RELEASES)
```

```python
totalCount = 0
mastersCount = 0
start_time = time.time()
```

```python
with codecs.open(pathMasters, "w", ENCODING) as mastersFH:
    mastersWriter = csv.writer(mastersFH, quoting=csv.QUOTE_MINIMAL)
 
    mastersWriter.writerow(['release_id','artist','genre','style', 'year','title','country', 'label', 'type','format'])
    
    for event, elem in etree.iterparse(pathWikiXML, events=('start', 'end')):
        tname = strip_tag_name(elem.tag)
        
        if event == 'start':
            if tname == 'release':
                release_id = elem.get("id")
                title = ""
                artist = ""
                country = ""
                disc_format = ""
                release_type = ""
                genre = []
                style = []
                year = None
        
        else:
            if tname == 'name' and elem.text and not artist:
                artist = elem.text

            elif tname == 'genre': #and elem.text and not genre:
                genre_text = elem.text
                genre.append(genre_text)
            
            elif tname == 'style' and elem.text and not style:
                style_text = elem.text
                style.append(style_text)
            
            elif tname == 'released':
                year = elem.text

            elif tname == 'country':
                country = elem.text
            
            elif tname == 'title' and elem.text and not title:
                title = elem.text
            
            elif tname == 'label':
                label = elem.get("name")
            
            elif tname == 'description' and elem.text and not release_type:
                release_type = elem.text
            
            elif tname == 'format':
                disc_format = elem.get("name")

            elif tname == 'release':
                totalCount += 1

                mastersCount += 1
                if country == "Portugal":
                    mastersWriter.writerow([release_id, artist, genre, style, year, title, country, label, release_type, disc_format])
                
                if totalCount > 1 and (totalCount % 100000) == 0:
                    print("{:,}".format(totalCount))

            elem.clear()
        
time_took = time.time() - start_time
print(f"Total runtime: {hms_string(time_took)}")
```
